```python
import numpy as np
from bokeh.plotting import output_notebook, figure, show
```

```python
x=np.linspace(-8*np.pi, 8*np.pi, 4000)
```

```python
type(x)
```

```python
y1=1
```

```python
y2=np.cos(x)
```

```python
y3=np.sin(x)/x
```

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"
```

```python
#p=figure(width=500)
#p=figure(plot_width=500,plot_height=500,title="Squeeze Theorem fixing sin(x)/x in place at 1", tools=TOOLS)
p=figure(title="Squeeze Theorem fixing sin(x)/x in place at 1", tools=TOOLS)
```

```python
#p.circle(x, y,size=2,  legend="sin(x)")
p.line(x, y1, legend="Y=1")
```

```python
p.line(x, y2, legend="cos(x)", line_color='orange')
```

```python
p.line(x, y3, legend="y = sinx(x)/x", line_color='purple')
```

```python
output_notebook()
```
